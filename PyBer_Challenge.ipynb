### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
ride_data_df["date"].max()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

# Create the Urban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]

urban_cities_df.head()

total_rides_urban= urban_cities_df.count()["ride_id"]
print(total_rides_urban)
```

```python
# Create the suburban city DataFrame and get the total rides for suburban city 
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]

suburban_cities_df.head()
suburban_cities_df.count()

total_rides_suburban= suburban_cities_df.count()["ride_id"]
print(total_rides_suburban)
```

```python
# Create the rural city DataFrame and gte the total rides for rural city

rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

rural_cities_df.head()
rural_cities_df.count()

total_rides_rural= rural_cities_df.count()["ride_id"]
print(total_rides_rural)
```

```python
# Get the total rides using group by type and count of ride id

total_rides_by_citytype = pyber_data_df.groupby(["type"]).count()["ride_id"]

total_rides_by_citytype
```

```python
# 2. Get the total drivers for each city type

# Get the driver count by city type

driver_count_by_citytype = city_data_df. groupby(["type"]).sum()["driver_count"]

driver_count_by_citytype
```

```python
#  3. Get the total amount of fares for each city type

# total_urbanfare = urban_cities_df.sum()["fare"]
# total_suburbanfare = suburban_cities_df.sum()["fare"]
# total_ruralfare = rural_cities_df.sum()["fare"]
# print(total_urbanfare)
# print(total_suburbanfare)
# print(total_ruralfare)

total_fare_by_citytype = pyber_data_df. groupby(["type"]).sum()["fare"]
total_fare_by_citytype
```

```python
#  4. Get the average fare per ride for each city type. 

#In Step 4, calculate the average fare per ride by city type by dividing the sum of all the fares by the total rides.

# avg_urbanfare = total_urbanfare/total_rides_urban
# avg_suburbanfare = total_suburbanfare/total_rides_suburban
# avg_ruralfare = total_ruralfare/total_rides_rural

# print(avg_urbanfare)
# print(avg_suburbanfare)
# print(avg_ruralfare)

average_fare_per_ride = total_fare_by_citytype/total_rides_by_citytype
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 

#In Step 5, calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers
average_fare_per_driver = total_fare_by_citytype/driver_count_by_citytype

average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides": total_rides_by_citytype,
                     "Total Drivers": driver_count_by_citytype,
                     "Total Fares": total_fare_by_citytype,
                     "Average Fare per Ride": average_fare_per_ride,
                     "Average Fare per Driver": average_fare_per_driver})

pyber_summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame

pyber_data_df
```

```python
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

totalfare_by_citytype_date = pyber_data_df.groupby(["type","date"]).sum()["fare"]
totalfare_by_citytype_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

totalfare_by_citytype_date = totalfare_by_citytype_date.reset_index()
totalfare_by_citytype_date
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

totalfare_by_citytype_date = totalfare_by_citytype_date.pivot(index = 'date',columns= 'type',values = 'fare')
totalfare_by_citytype_date
```

```python
totalfare_by_citytype_date.info()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_filter_df = totalfare_by_citytype_date.loc['2019-01-01':'2019-04-29']
date_filter_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

date_filter_df.index = pd.to_datetime(date_filter_df.index)
date_filter_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_filter_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = date_filter_df.resample("W").sum()
weekly_fares_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
weekly_fares_df.plot(figsize = (15,10))

#Annotate the y-axis label and the title, 
#then use the appropriate code to save the figure as PyBer_fare_summary.png in your "analysis" folder.

plt.title ("Total fare by City Type")
plt.xlabel("Months")
plt.ylabel("Fare ($USD)")

plt.legend()

plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
